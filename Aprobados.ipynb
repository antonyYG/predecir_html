```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
!pip install "keras==2.15.0" tensorflowjs
```

```python
datos = {
    'Alumnos': np.arange(1,51),
    'Asistencias': np.random.randint(1,50,50),
    'Faltas': np.random.randint(1,50,50)
}

df = pd.DataFrame(datos)
df
```

```python
df['Abandono'] = np.where((df['Faltas'] > 8) & (df['Asistencias']< 20), 1, 0)
df
```

```python
X = df[['Asistencias', 'Faltas']].values
y = df['Abandono'].values
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=4, activation="relu", input_shape=(2,)),  # capa oculta
    tf.keras.layers.Dense(units=1, activation="sigmoid")  # salida binaria
])
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
history = model.fit(X, y, epochs=50, verbose=0)
```
